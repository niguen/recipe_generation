```python
import pandas as pd
import os
import ast
```

```python
mypath = "."
folder_name='archive/'
file_path = f'{mypath}/{folder_name}/'

filename = os.path.join(file_path, 'RAW_recipes.csv')
RAW_recipes_df = pd.read_csv(filename)
RAW_recipes_df.head(5)
```

```python
tags = set()
all_tags = []
for list in RAW_recipes_df['tags'].to_list():
    current_list = ast.literal_eval(list)
    tags.update(current_list)
    all_tags.append(current_list)
```

```python
all_tags = [item for sublist in all_tags for item in sublist]
```

```python
tag_map = []
for tag in tags:
    occur = all_tags.count(tag)
    tag_map.append({'tag': tag, 'n_times':occur})
```

```python
df = pd.DataFrame(tag_map)
df.sort_values(by=['n_times'], inplace=True, ascending=False)
df.head(30)
```

```python
pattern = 'vegetables'
is_included = RAW_recipes_df.tags.str.contains(pattern)
RAW_recipes_df['is_included'] = is_included
is_included.value_counts()
```

```python
selected_recipes = RAW_recipes_df.loc[RAW_recipes_df['is_included']]
selected_recipes.shape
```

## Ingredient Diversity

```python
#ingredient_set = set()
ingredient_list = []
for list in selected_recipes['ingredients'].to_list():
    current_list = ast.literal_eval(list)
    for item in current_list:
        #print(item)
        #ingredient_set.update(item)
        ingredient_list.append(item)
```

```python
ingredient_list
```

```python
ingredient_set = set(ingredient_list)
```

```python
ingredient_map = []
for ingredient in ingredient_set:
    occur = ingredient_list.count(ingredient)
    ingredient_map.append({'ingredient': ingredient, 'n_times':occur})
```

```python
ingredient_df = pd.DataFrame(ingredient_map)
ingredient_df.sort_values(by=['n_times'], inplace=True, ascending=True)
ingredient_df.shape
```
